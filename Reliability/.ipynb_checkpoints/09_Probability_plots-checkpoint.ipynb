# PROBABILITY PLOTS

```python
import matplotlib.pyplot as plt
from reliability.Distributions import Normal_Distribution, Exponential_Distribution, Weibull_Distribution
from reliability.Probability_plotting import Normal_probability_plot, Exponential_probability_plot, Weibull_probability_plot
from reliability.Other_functions import make_right_censored_data
```

# Example 1

```python
dist = Normal_Distribution(mu = 50, sigma = 10)
```

```python
failures = dist.random_samples(100, seed = 5)
```

```python
Normal_probability_plot(failures = failures)
dist.CDF(linestyle='--',label='True CDF')
plt.legend()
plt.show()
```

## Example 2

```python
dist = Exponential_Distribution(Lambda = 0.25, gamma = 12)
```

```python
raw_data = dist.random_samples(100, seed = 42)
```

```python
data = make_right_censored_data(raw_data, threshold = 17)
```

```python
Exponential_Distribution(Lambda = 0.25).CDF(linestyle='--', label='True CDF')
Exponential_probability_plot(
    failures = data.failures, 
    right_censored = data.right_censored,
    fit_gamma = True
)
plt.legend()
plt.show()
```

# Example 3

```python
dist = Weibull_Distribution(alpha = 250, beta = 3)
```

```python
for i, x in enumerate([10, 100, 1000]):
    plt.subplot(131 + i)
    dist.CDF(linestyle='--', label='True CDF')
    failures = dist.random_samples(x, seed = 42)
    Weibull_probability_plot(failures = failures)
    plt.title(str(str(x) + 'samples'))
plt.gcf().set_size_inches(15, 7)
plt.tight_layout()
plt.show()
```
