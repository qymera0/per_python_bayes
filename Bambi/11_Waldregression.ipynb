# WALD ANG GAMMA REGRESSION

## Load libraries

```python
import arviz as az
import bambi as bmb
import matplotlib.pyplot as plt
import numpy as np
```

```python
az.style.use("arviz-darkgrid")
np.random.seed(1234)
```

## Load data

```python
data = bmb.load_data("carclaims")
data.head()
```

```python
data = (data
        .query("claimcst0 > 0")
)
```

```python
plt.hist(data[data["claimcst0"] <= 15000]["claimcst0"], bins=30)
plt.title("Distribution of claim amount")
plt.xlabel("Claim amount ($)");
```

## Wald model

```python
modelWald = bmb.Model(
                        "claimcst0 ~ C(agecat) + gender + area",
                        data,
                        family = 'wald',
                        link = 'log'
)
```

```python
modelWald.build()
modelWald.graph()
```

```python
fitted_wald = modelWald.fit(tune=2000, target_accept=0.9, idata_kwargs={"log_likelihood": True})
```

```python
az.plot_trace(fitted_wald);
```

```python
az.summary(fitted_wald)
```

## Gamma regression

```python
modelGamma = bmb.Model(
                        "claimcst0 ~ agecat + gender + area",
                        data,
                        family = 'gamma',
                        link = 'log',
                        categorical = 'agecat'
)
```

```python
modelGamma.build()
modelGamma.graph()
```

```python
fitted_gamma = modelGamma.fit(tune=2000, target_accept=0.9, idata_kwargs={"log_likelihood": True})
```

```python
az.plot_trace(fitted_gamma);
```

## Model comparison

```python
models = {
    'wald': fitted_wald,
    'gamma': fitted_gamma
}
```

```python
df_compare = az.compare(models)
```

```python
df_compare
```

```python
az.plot_compare(df_compare, insample_dev = False)
```
