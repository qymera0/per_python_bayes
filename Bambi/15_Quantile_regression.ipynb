# QUANTILE REGRESSION

## Load libraries

```python
import arviz as az
import bambi as bmb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import pymc as pm
```

```python
az.style.use("arviz-darkgrid")
SEED = 12947
```

## Load data

```python
data = pd.read_csv(pm.get_data("bmi.csv"))
```

```python
data.head()
```

## Plot data

```python
plt.plot(data.age, data.bmi, "k.");
plt.xlabel("Age")
plt.ylabel("BMI");
```

## Data transformation

```python
# From q to Kappa
quantiles = np.array([0.1, 0.5, 0.9])
kappas = (quantiles/(1-quantiles))**0.5
```

```python
knots = np.quantile(data.age, np.linspace(0, 1, 10))[1:-1]
```

```python
knots
```

## Model

```python
idatas = []
```

```python
for k in kappas:
    model = bmb.Model(
                        'bmi ~ bs(age, knots = knots)', # Using a spline
                        data = data,
                        family = "asymmetriclaplace",
                        priors = {
                            'kappa': k
                        }
    )
    idata = model.fit()
    model.predict(idata)
    idatas.append(idata)
```

```python
plt.plot(data.age, data.bmi, ".", color="0.5")
for idata, q in zip(idatas, quantiles):
    plt.plot(data.age.values, idata.posterior["bmi_mean"].mean(("chain", "draw")),
            label=f"q={q:}", lw=3);
    
plt.legend()
plt.xlabel("Age")
plt.ylabel("BMI");
```
